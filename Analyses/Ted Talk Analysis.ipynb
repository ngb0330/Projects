```python
ted_data['tags_cleaned'] = ted_data['tags'].str.strip('][').str.replace("''","").str.split(',')
```

```python
[tag for tag in ted_data['tags_cleaned'][3]]
```

```python
tags_list = []
for i in range(ted_data.shape[0]):
    for tag in ted_data['tags_cleaned'][i]:
        if tag not in tags_list:
            tags_list.append(tag)
```

```python
sorted(tags_list)
```

```python
for tag in tags_list:
    if tag in ted_data['tags_cleaned'][0]:
        print(tag + 'one')
    else:
        print(tag + 'zero')
```

```python
def add_dummy_tags(original_data):
    data = original_data.copy()
    tags_list = []
    for i in range(data.shape[0]):
        for tag in data['tags_cleaned'][i]:
            if tag not in tags_list:
                tags_list.append(tag)
    for tag in tags_list[1:]:
        data['has_{}_tag'.format(tag.replace("''", ""))] = int(tag in ted_data['tags_cleaned'])
    return data
```

```python
ted_data_tag_dummies = add_dummy_tags(ted_data)
```
