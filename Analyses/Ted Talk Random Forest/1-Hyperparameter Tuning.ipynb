```python
X_train = pd.read_feather('with_dates_X_train')
y_train = pd.read_feather('with_dates_y_train')
```

```python
X_train = X_train.drop(columns = 'index')
y_train = y_train.drop(columns = 'index')
```

```python
X_val = pd.read_feather('with_dates_X_validate')
y_val = pd.read_feather('with_dates_y_validate')
```

```python
X_val = X_val.drop(columns = 'index')
y_val = y_val.drop(columns = 'index')
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators = 100,random_state = 35)
rfr.fit(X_train, y_train)
```

```python
import numpy as np
tree_predictions = np.vstack([estimator.predict(X_val) for estimator in rfr.estimators_])
```

```python
from sklearn.metrics import r2_score
r2_scores = [r2_score(y_val, np.mean(tree_predictions[:i + 1], axis = 0)) for i in range(0, 100)]
x_vals = list(range(1, 100))
```

```python
from graph import trace_values, plot
trace = trace_values(x_vals, r2_scores)
plot([trace])
```

Because my scores seem to stabilize around 21 estimators, I choose 21 to be my n_estimators.

```python
rfr_list = []
for feature in ['auto', 'sqrt', 'log2', None, .5]:
    rfr = RandomForestRegressor(n_estimators = 21, max_features = feature, random_state = 35)
    rfr_list.append([feature, rfr.fit(X_train, y_train).score(X_val, y_val)])
```

```python
rfr_list
```

Because the model performs best with a a max_features value of auto, I choose that to be my max_features value. <br>
This means that the number of features looked at when performing a split is the number of features in the dataset. <br>

```python
min_samples = [integer for integer in range(2,100)]
scores = []
for integer in min_samples:
    rfr = RandomForestRegressor(n_estimators = 21, max_features = 'auto', min_samples_leaf = integer, random_state = 35)
    scores.append([integer, rfr.fit(X_train, y_train).score(X_val, y_val)]);
```

```python
wrapped_scores = np.array(scores)
```

```python
wrapped_scores[:5]
```

```python
trace= trace_values(wrapped_scores[:, 0], wrapped_scores[:, 1])
plot([trace])
```

Because there is a steady decline in the scores for the random forest after a min_samples_leaf value of 5, I chose 5 to be my min_samples leaf value.

So my Random Forest Regressors will have 21 estimators, with 'auto' max_features (meaning all features are considered when looking for the best split), and a minimum of 5 samples per leaf.
